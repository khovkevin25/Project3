```python
import pandas as pd
import altair as alt

# Importing the data
data_path = "https://khovkevin25.github.io/Project3/static/Data/Philadelphia_Food_Access.csv"
df = pd.read_csv(data_path)

# Remove blank rows
median_updated = df.dropna(subset=["Median Family Income"])

# Create scatter plot
scatter = alt.Chart(median_updated).mark_circle(size=70, opacity=0.6, color="purple").encode(
    x= alt.X("Median Family Income", title="Median Family Income", scale=alt.Scale(domain=[0, 240000])),
    y= alt.Y("Snap Percentage", title="% of Households on SNAP", scale=alt.Scale(domain=[0, 80])),
    tooltip=["Census Tract", "Median Family Income", "Snap Percentage"]
).properties(
    title= "Correlation: Median Family Income and SNAP Percentage",
    width= 800,
    height= 800
)

# Add regression line
regress_line = scatter.transform_regression("Median Family Income", "Snap Percentage").mark_line(color="orange")

# Print scatter plot
scatter + regress_line
```
