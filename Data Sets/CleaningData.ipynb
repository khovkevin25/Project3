```python
import pandas as pd
```

```python
df = pd.read_excel("FoodAccess.xlsx", sheet_name = 'Food Access Research Atlas')
```

```python
df.head()
```

```python
Cleaned = df[["CensusTract","Pop2010", "OHU2010", "LILATracts_1And10", "LILATracts_halfAnd10", "LowIncomeTracts", "PovertyRate", "MedianFamilyIncome", "LA1and10", "LAhalfand10", "lalowihalf", "lakidshalf", "laseniorshalf", "lakids1", "laseniors1","lasnaphalf","lasnap1", "TractSNAP"]]

```

```python
Cleaned.head()
```

```python
SuperCleaned = Cleaned.rename(columns={"CensusTract":"Census Tract", "Pop2010": "Population", "OHU2010": "Housing Units in Tract", "LILATracts_1And10": "LI, LA - 1 mile radius", "LILATracts_halfAnd10": "LI, LA - 0.5 mile radius", "LowIncomeTracts": "Low Income Tracts", "PovertyRate": "Poverty Rate (%)", "MedianFamilyIncome": "Median Family Income", "LA1and10": "LA - 1 mile radius", "LAhalfand10": "LA - 0.5 mile radius", "lalowihalf": "LA,LI - 0.5 mile radius", "lakidshalf": "LA - Kids population, 0.5 mile radius", "laseniorshalf": "LA - Senior population, 0.5 mile radius","lakids1": "LA - Kids population, 1 mile radius", "laseniors1": "LA - Senior population, 1 mile radius", "lasnaphalf": "LA - Housing Units w/ Snap Benefits, 0.5 mile radius","lasnap1": "LA - Housing Units w/ Snap Benefits, 1 mile radius", "TractSNAP": "Total Count of Housing w/ Snap Benefits"})
```

```python
SuperCleaned.head()
```

```python
SuperCleaned['Poverty Rate (%)'] = SuperCleaned['Poverty Rate (%)'].astype(float)
```

```python
SuperCleaned.style.format({'Poverty Rate (%)': '{:,.2f}'.format, 'Median Family Income': '{:,.2f}'.format, 'LA,LI - 0.5 mile radius': '{:,.2f}'.format, 'LA - Kids population, 0.5 mile radius': '{:,.2f}'.format, 'LA - Senior population, 0.5 mile radius': '{:,.2f}'.format, 'LA - Kids population, 1 mile radius': '{:,.2f}'.format, 'LA - Senior population, 1 mile radius': '{:,.2f}'.format,'LA - Housing Units w/ Snap Benefits, 0.5 mile radius': '{:,.2f}'.format, 'LA - Housing Units w/ Snap Benefits, 1 mile radius': '{:,.2f}'.format})
```

```python
SuperCleaned.info()
```

```python
SuperCleaned.to_excel("Philadelphia Food Access.xlsx", index = False)
```
